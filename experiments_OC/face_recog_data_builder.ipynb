```python
fvc_root_dir = '/media/data1/FaceForensics_Dec2020/FaceForensics++/face_data/data_raw/original_sequences/youtube/raw/faces/*' 

#'/media/data1/FaceForensics_Dec2020/FaceForensics++/face_data/data_raw/original_sequence/youtube/raw/faces/'
fvc_root_list = sorted(glob.glob(fvc_root_dir)) #sorted(os.listdir(fvc_root_dir))
# print(fvc_root_list)

exp = '.png'

full_train_list = []
min_num = 1000
for root_path in fvc_root_list:
    # print(root_path)
    sub_list = sorted(glob.glob(root_path+f"/*00{exp}"))
    sub_list = sub_list + sorted(glob.glob(root_path+f"/*30{exp}"))
    sub_list = sub_list + sorted(glob.glob(root_path+f"/*60{exp}"))
    # print(len(sub_list))
    full_train_list += sub_list[:9]
    
    if len(sub_list) < min_num:
        min_num = len(sub_list)

print(min_num, len(full_train_list))
```

```python
fvc_root_dir = '/media/data1/FaceForensics_Dec2020/FaceForensics++/face_data/data_raw/original_sequences/youtube/raw/faces/*' 

#'/media/data1/FaceForensics_Dec2020/FaceForensics++/face_data/data_raw/original_sequence/youtube/raw/faces/'
fvc_root_list = sorted(glob.glob(fvc_root_dir)) #sorted(os.listdir(fvc_root_dir))
# print(fvc_root_list)

exp = '.png'

full_test_list = []
min_num = 1000
for root_path in fvc_root_list:
    # print(root_path)
    sub_list = sorted(glob.glob(root_path+f"/*80{exp}"))
    # print(len(sub_list))
    full_test_list += sub_list[:3]
    
    if len(sub_list) < min_num:
        min_num = len(sub_list)

print(min_num, len(full_test_list))
```
